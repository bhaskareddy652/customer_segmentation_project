```python
import os
os.environ["OMP_NUM_THREADS"] = "1"
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score, silhouette_score
```

```python
df1=pd.read_csv("Downloads/Customers.csv")
df2=pd.read_csv("Downloads/Products.csv")
df3=pd.read_csv("Downloads/Transactions.csv")
```

```python
def data_desc(df, name):
    print(f"\n--- {name} data set Overview ---")
    print(df.info())
    print(df.describe(include='all'))
    print(f"Missing Values:\n{df.isnull().sum()}")
```

```python
data_desc(df1, "Customers")
data_desc(df2, "Products")
data_desc(df3, "Transactions")
```

```python
data = pd.merge(df3,df1, on="CustomerID", how="inner")
data = pd.merge(data, df2, on="ProductID", how="inner")
```

```python
total_sales = data["TotalValue"].sum()
unique_customers = data["CustomerID"].nunique()
unique_products = data["ProductID"].nunique()
total_transactions = data["TransactionID"].nunique()
```

```python
print("\n--- Key Metrics ---")
print(f"Total Sales: ${total_sales:.2f}")
print(f"Total Transactions: {total_transactions}")
print(f"Unique Customers: {unique_customers}")
print(f"Unique Products: {unique_products}")
```

```python
print("\n--- Top-Selling Products ---")
top_products = data.groupby("ProductName")["TotalValue"].sum().sort_values(ascending=False).head(5)
top_products
```

```python
print("\n--- Top Regions by Sales ---")
top_regions = data.groupby("Region")["TotalValue"].sum().sort_values(ascending=False)
top_regions
```

```python
print("\n--- Customer Signup Trends ---")
df1["SignupDate"] = pd.to_datetime(df1["SignupDate"])
customer_signup_trends = df1.groupby(df1["SignupDate"].dt.to_period("M")).size()
customer_signup_trends
```

```python
plt.figure(figsize=(10, 6))
top_products.plot(kind="bar", title="Top-Selling Products", xlabel="Product", ylabel="Total Sales ($)", color="blue")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
top_regions.plot(kind="bar", title="Top-Regions by Sales", xlabel="Region", ylabel="TotalValue", color="orange")
plt.xticks(rotation=45)
plt.show()
```
